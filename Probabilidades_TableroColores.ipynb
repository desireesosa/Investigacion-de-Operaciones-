*Definicion:* *Una Cadena de Markov es un proceso estocástico (aleatorio) que modela un sistema que cambia de un estado a otro en una secuencia de pasos o intervalos de tiempo.*

*En términos simples, en una cadena de Markov, el estado futuro del sistema se puede predecir solo conociendo su estado presente, lo que implica que no hay memoria de eventos pasados.*

<font color="fiusca">Ejemplo de una Cadena de Markov:</font>

*Se representa mediante una matriz de transición, donde cada entrada indica la probabilidad de moverse de un estado a otro.*

*Para la solución de estos problemas existe el **analisis analitico mediante cadenas de Markov**, existe una variable de estado que representa la posición actual de la ficha.*


*   ¿Qué es? Es un vector colunma que contiene las robabilidades de que el sistema se encuentre en cada uno de los posibles estados en un momento específico (n).

* Propiedad Clave: La suma de todos los elementos dentro del vector de estado siempre debe ser 1 (o 100%), ya que representa la distribución total de probabilidad en ese paso.

Programa para calcular las probabilidades x_n, y_n, z_n, t_n y graficarlas.

En nuestro problema:
###

| Elemento | Símbolo | Descripción |
| :--- | :--- | :--- |
| **Estados del Sistema** | $x, y, z, t$ | Estos son los **cuatro posibles sectores** (o colores) en los que se puede encontrar el elemento móvil. |
| **Variable de Estado** | $\mathbf{X}_n = \begin{pmatrix} x_n \\ y_n \\ z_n \\ t_n \end{pmatrix}$ | **Vector de Probabilidades** en el paso $n$. |
| **Significado de $x_n$** | $x_n$ | Es la **probabilidad** de que el elemento se encuentre en el **Sector $X$** después de $n$ transiciones. (Aplica igual para $y_n, z_n, t_n$). |

```python
#Aplicando el ejercicio en clase:
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Matriz de Transición.
A = np.array([
    [1/4, 0/4, 1/4, 2/4],
    [2/4, 1/4, 0/4, 1/4],
    [1/4, 2/4, 1/4, 0/4],
    [0/4, 1/4, 2/4, 1/4]
])

print("Matriz de Transición A:")
print(A)
```

```python
#definir los pasos, las probabilidades, etc.
#iniciamos en sector x
X_0 = np.array([[1.0], [0.0], [0.0], [0.0]])
n_steps = 50 #pasos


#transiciones
x_history = [X_0[0, 0]]
y_history = [X_0[1, 0]]
z_history = [X_0[2, 0]]
t_history = [X_0[3, 0]]

X_n = X_0
```

```python
for n in range(1, n_steps + 1):
    # Cálculo del nuevo vector de estado: X_n = A * X_{n-1}
    X_n = A @ X_n

    # Almacenamiento de los nuevos valores
    x_history.append(X_n[0, 0])
    y_history.append(X_n[1, 0])
    z_history.append(X_n[2, 0])
    t_history.append(X_n[3, 0])
```

```python
#grafica del tablero de colores
time_steps = np.arange(n_steps + 1)
plt.figure(figsize=(10, 6))

plt.plot(time_steps, x_history, label='$x_n$', marker='.', linestyle='-')
plt.plot(time_steps, y_history, label='$y_n$', marker='.', linestyle='-')
plt.plot(time_steps, z_history, label='$z_n$', marker='.', linestyle='-')
plt.plot(time_steps, t_history, label='$t_n$', marker='.', linestyle='-')

plt.title('Evolución de las Probabilidades')
plt.xlabel('Paso de Tiempo')
plt.ylabel('Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(y=0.25, color='r', linestyle=':', label='Límite')
plt.ylim(0, 1.05)
plt.show()
```
